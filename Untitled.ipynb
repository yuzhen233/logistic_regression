```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import seaborn as sns

from scipy.optimize import minimize

from sklearn.preprocessing import PolynomialFeatures

%matplotlib inline 
```

# `1)逻辑斯特回归`

```python
data = np.loadtxt(fname='data1.txt', delimiter=',')
# 读取数据前五行
print(data[0:5, :])
```

```python
# 描述数据
def plotData(data, xlabel, ylabel):
    
    neg = (data[:, 2] == 0)
    pos = (data[:, 2] == 1)

    plt.scatter(x=data[neg][:, 0], y=data[neg][:, 1])
    plt.scatter(x=data[pos][:, 0], y=data[pos][:, 1])
    
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
plotData(data, 'score1', 'score2')
```

```python
X = np.c_[np.ones((data.shape[0],1)), data[:,0:2]]
y = np.c_[data[:,2]]
theta = np.zeros(X.shape[1])
```

# `1.1)损失函数`

```python
#定义sigmoid函数
def sigmoid(z):
    return(1 / (1 + np.exp(-z)))
```

```python
#定义损失函数
def costFunction(theta, X, y):
    m = y.size
    h = sigmoid(X.dot(theta))
    
    J = -1.0*(1.0/m)*(np.log(h).T.dot(y)+np.log(1-h).T.dot(1-y))
               
    if np.isnan(J[0]):
        return(np.inf)
    return J[0]
```

# `1.2)梯度下降`

```python
#求解梯度
def gradient(theta, X, y):
    m = y.size
    h = sigmoid(X.dot(theta.reshape(-1,1)))
    
    grad =(1.0/m)*X.T.dot(h-y)

    return(grad.flatten())
```

```python
res = minimize(costFunction, theta, args=(X, y), jac=gradient, options={'maxiter':400})
```

```python
theta_result = res.x
print(theta_result)
```

# `1.3)预测`

```python
test_x = np.array([1, 65, 65])
print(sigmoid(np.dot(test_x, theta_result)))
```

# `2)添加正则化的逻辑斯特回归`

```python
data2 = np.loadtxt('data2.txt', delimiter=',')
print(data2.shape)
print(data2[0:5, :])
```

```python
plotData(data2, 'test1', 'test2')
```

```python
poly = PolynomialFeatures(6)
X = data2[:,0:2]
print(X[0:5, :])
XX = poly.fit_transform(data2[:,0:2])
print(XX[0:1, :])
y = data2[:,2]
# 看看形状(特征映射后x有多少维了)
print(XX.shape)
y = np.c_[data2[:, 2]]
theta = np.zeros(XX.shape[1])
```

# `2.1)损失函数`

```python
def computCostReg(theta, reg, XX, y):
    m = y.size
    h = sigmoid(XX.dot(theta))
    
    J = -1.0*(1.0/m)*(np.log(h).T.dot(y)+np.log(1-h).T.dot(1-y)) + (reg/(2.0*m))*np.sum(np.square(theta[1:]))
    
    if np.isnan(J[0]):
        return(np.inf)
    return(J[0])
```

```python
computCostReg(theta, 1, XX, y)
```

# `2.2)梯度下降`

```python
def gradientReg(theta, reg, *args):
    m = y.size
    h = sigmoid(XX.dot(theta.reshape(-1,1)))
      
    grad = (1.0/m)*XX.T.dot(h-y) + (reg/m)*np.r_[[[0]],theta[1:].reshape(-1,1)]
        
    return(grad.flatten())
```

```python
res2 = minimize(computCostReg, theta, args=(1.0, XX, y), jac=gradientReg, options={'maxiter':3000})
print(res2.x)
```

# `2.3)预测`

```python
test_xx = poly.fit_transform(np.array([[0.25, -0.75]]))
print(sigmoid(np.dot(test_xx, res2.x)))
test_xx = poly.fit_transform(np.array([[0.75, 0.50]]))
print(sigmoid(np.dot(test_xx, res2.x)))
test_xx = poly.fit_transform(np.array([[0.25, 0.75]]))
print(sigmoid(np.dot(test_xx, res2.x)))
```

# `2.4)绘制决策边界`

```python
plotData(data2, 'test1', 'test2')
plt.scatter(0.25, -0.75, marker='v', c='k')
plt.scatter(0.25, 0.75, marker='+', c='k')

x1_min, x1_max = X[:,0].min(), X[:,0].max(),
x2_min, x2_max = X[:,1].min(), X[:,1].max(),
xx1, xx2 = np.meshgrid(np.linspace(x1_min, x1_max), np.linspace(x2_min, x2_max))
h = sigmoid(poly.fit_transform(np.c_[xx1.ravel(), xx2.ravel()]).dot(res2.x))
h = h.reshape(xx1.shape)
plt.contour(xx1, xx2, h, [0.5], linewidths=1, colors='y');
```
